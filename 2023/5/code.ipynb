```python
class RangeMap:
    def __init__(self, src, dest, size):
        self.src = src
        self.dest = dest
        self.size = size

    def __contains__(self, item):
        return self.src <= item < self.src + self.size

    def __getitem__(self, item):
        if item not in self:
            return item
        return self.dest + item - self.src

    def __repr__(self):
        return f"RangeMap({self.src}, {self.dest}, {self.size})"
```

```python
class Range:
    def __init__(self, src, size):
        self.src = src
        self.size = size

    def __repr__(self) -> str:
        return f"Range({self.src}, {self.size})"

    def __contains__(self, item):
        return self.src <= item < self.src + self.size

    def __getitem__(self, item):
        return self.src + item
```

```python
data = [x.split(":;") for x in ";".join(data).split(";;")]
seeds, data = list(map(int, data[0][0].split(": ")[1].split(" "))), data[1:]
maps = {i: {"data": np.array(x[1].split(";"))} for i, x in enumerate(data)}

rprint(seeds, maps)
```

```python
for v in maps.values():
    v["map"] = []
    for i in v["data"]:
        dest, src, span = map(int, i.split(" "))
        v["map"].append(RangeMap(src, dest, span))


rprint(maps)
```

```python
locations = []
for seed in seeds:
    for map_ in maps.values():
        for i in map_["map"]:
            if seed in i:
                seed = i[seed]
                break
        else:
            seed = seed

    locations.append(seed)

print(f"Part 1: {min(locations)}")
```
